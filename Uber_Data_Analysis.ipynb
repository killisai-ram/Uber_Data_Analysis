```python
# Reading the Uber dataset into a pandas DataFrame
# This line loads data from the specified CSV file, which should be located in the working directory or the provided path.
dataset=pd.read_csv("UberDataset.csv")
```

```python
dataset
```

```python
# Displaying the shape of the dataset
# The shape attribute returns a tuple representing the dimensions of the DataFrame.
# It outputs (number_of_rows, number_of_columns), which helps understand the size of the dataset.
dataset.shape
```

```python
# Displaying a summary of the dataset
dataset.info()
```

```python
# Filling missing values in the 'PURPOSE' column
# The fillna() method replaces all NaN (missing) values in the 'PURPOSE' column with the string "NOT".
# The inplace=True parameter ensures the operation is performed directly on the dataset without needing reassignment.

dataset['PURPOSE'].fillna("NOT",inplace=True)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
# Converting 'START_DATE' and 'END_DATE' columns to datetime format
# The pd.to_datetime() function converts the data in these columns to pandas datetime objects.
# - errors='coerce' ensures that any invalid date formats are replaced with NaT (Not a Time) instead of raising an error.
# This standardizes the date format, enabling proper datetime operations like filtering, sorting, or calculating durations.

dataset['START_DATE']= pd.to_datetime(dataset['START_DATE'],errors='coerce')

dataset['END_DATE']= pd.to_datetime(dataset['END_DATE'],errors='coerce')
```

```python
dataset.info()
```

```python
from datetime import datetime
# Extracting the date part from the 'START_DATE' column
# The pd.DatetimeIndex() function accesses the datetime object for each entry in the 'START_DATE' column.
# The .date attribute retrieves only the date (year, month, day) and assigns it to the new 'date' column.

dataset['date']=pd.DatetimeIndex(dataset['START_DATE']).date

# Extracting the hour (time part) from the 'START_DATE' column
# The .hour attribute retrieves the hour component (0-23) of the datetime object, representing the start time of the trip.
# This is assigned to a new column named 'time'.

dataset['time']=pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head()
```

```python
# Categorizing the time of day into day parts: morning, afternoon, evening, and night
# The pd.cut() function bins the 'time' column values into specified intervals (bins):
# - bins=[0, 10, 15, 19, 24] defines the time ranges:
#   - 010: morning
#   - 1015: afternoon
#   - 1519: evening
#   - 1924: night
# - labels=['morning', 'afternoon', 'evening', 'night'] assigns descriptive labels to each bin.
# This creates a new 'day-night' column indicating the part of the day for each entry.

dataset['day-night']=pd.cut(x=dataset['time'],bins=[0,10,15,19,24],labels=['morning','afternoon','evening','night'])
```

```python
dataset.head()
```

```python
# Dropping rows with missing values
# The dropna() method removes any rows in the dataset that contain NaN or missing values in any column.
# The inplace=True parameter modifies the original dataset directly, rather than returning a new DataFrame.

dataset.dropna(inplace=True)
dataset.shape
```

Data Visualization

```python
# Setting up the figure size for the plots
# The figsize=(20, 5) creates a large figure with a width of 20 inches and a height of 5 inches.
plt.figure(figsize=(20,5))

# Creating the first subplot for 'CATEGORY' column
# plt.subplot(1, 2, 1) means we're creating a grid of 1 row and 2 columns, and this plot will occupy the first slot.
plt.subplot(1,2,1)


sns.countplot(dataset['CATEGORY'])
plt.xticks(rotation=90)
plt.title('Category')

plt.subplot(1,2,2)
sns.countplot(dataset['PURPOSE'])

# Displaying the plots
plt.show()
```

```python
#  PURPOSE

from matplotlib import pyplot as plt
import seaborn as sns
dataset.groupby('PURPOSE').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
sns.countplot(dataset['day-night'])
```

```python
dataset['MONTH']=pd.DatetimeIndex(dataset['START_DATE']).month

month_label ={1.0: 'Jan',2.0:'Feb',3.0:'Mar',4.0:'Apr',5.0:'May',6.0:'Jun',7.0:'Jul',8.0:'Aug',9.0:'Sep',10.0:'Oct',11.0:'Nov',12.0:'Dec'}

dataset['MONTH']=dataset['MONTH'].map(month_label)

mon=dataset.MONTH.value_counts(sort=False)
```

```python
dataset.head()
```

```python
df=pd.DataFrame({'Month':mon.values,
                 'Value Count':dataset.groupby('MONTH',sort=False)['MILES'].max()
})

p=sns.lineplot(data=df)
p.set(xlabel='Month',ylabel='Value Count')
```

```python
dataset['Day']=dataset.START_DATE.dt.weekday

day_label={0.0:'Monday',1.0:'Tuesday',2.0:'Wednesday',3.0:'Thursday',4.0:'Friday',5.0:'Saturday',6.0:'Sunday'}

dataset['Day']=dataset['Day'].map(day_label)
```
